```python
import modin.pandas as pd
import numpy as np

import pickle

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
import keras.backend as K
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
training_dat = pd.read_pickle('./PROCESSED/training_dat2.pkl')
training_dat.columns
```

```python
with open("./PROCESSED/scalers.pkl","rb") as f:
    scaler_dct = pickle.load(f)
scaler_dct.keys()
```

Converting scaled values back to original for analysis

```python
scaled_cols = list(scaler_dct.keys())
for col in scaled_cols:
    scaler = scaler_dct[col]
    training_dat[col] = scaler.inverse_transform(training_dat[[col]])
training_dat.head()
```

## Compare credit scores of non delinquent vs delinquent owners

```python
tf = training_dat['Loan Deliquency Within Year'] == 0
ax = sns.histplot(x="Credit Score", data=training_dat[tf], kde=True)
ax.set_title("Distribution With No Delinquency")
```

```python
tf = training_dat['Loan Deliquency Within Year'] == 1
ax = sns.histplot(x="Credit Score", data=training_dat[tf], kde=True)
ax.set_title("Distribution With Delinquency")
```

Buyers that suffer a loan delinquency within a year tend 
are more likely to have lower credit scores then the general 
loan population.

## Checkout Mortgage Insurance Percentage (MI %)

```python
mi_gp = training_dat.groupby('Mortgage Insurance Percentage (MI %)')
mi_dat = mi_gp.agg({'Loan Deliquency Within Year':'mean', 
                    'Loan Sequence Number':'count'})
# only look at buckets with more than 1000 loans
tf = mi_dat['Loan Sequence Number'] > 1000
mi_dat[tf]
```

Change this to boolean feature, since % value doesn't seem significant

```python
training_dat['No Mortgage Insurance'] = 1

tf = training_dat['Mortgage Insurance Percentage (MI %)'] > 0
training_dat.loc[tf, 'No Mortgage Insurance'] = 0
```

```python
training_dat.groupby('No Mortgage Insurance')['Loan Deliquency Within Year'].mean()
```

There is a stronger likelyhood of delinquency on loans with mortgage insurance.

## Binary flags

```python
binary_cols = ['Units_1', 'Units_2', 'Units_3', 'Units_4', 
               'OccupancyStatus_I', 'OccupancyStatus_P', 'OccupancyStatus_S', 
               'Channel_B', 'Channel_C', 'Channel_R', 
               'PropertyType_CO', 'PropertyType_CP', 'PropertyType_MH', 'PropertyType_PU', 'PropertyType_SF',
                'LoanPurpose_C', 'LoanPurpose_N', 'LoanPurpose_P', 
               'LoanTerm_360', 'LoanTerm_180', 'LoanTerm_240', 'LoanTerm_120', 'LoanTerm_300',
               'OneBorrower', 'AffordableProgramFlag']
for col in binary_cols:
    tf = training_dat[col] == 1
    drat = training_dat[tf]['Loan Deliquency Within Year'].sum() / training_dat[col].sum()
    
    print("Delinquency rate for " + str(col) + ": " + str(drat))
    print(str(col) + " count: " + str(training_dat[col].sum()))

drat = training_dat['Loan Deliquency Within Year'].sum() / len(training_dat)
print("Overall delinquency: " + str(drat))
print("Total count: " + str(len(training_dat)))
```

I should get rid of PropertyType_CP due to few observations, but otherwise these flags have valuable information.
