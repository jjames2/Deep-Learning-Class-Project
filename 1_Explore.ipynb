```python
month_df = pd.read_csv('./DATA/sample_svcg_2016.txt',sep='|',names=list(col_names['Svcg']))
month_df.head()
```

```python
tf1 = month_df['Loan Age'] < 13
tf2 = month_df['Delinquency Due to Disaster'] != 'Y'
tf3 = ~month_df['Current Loan Delinquency Status'].isin(['R','XX','   '])
tf = tf1 & tf2 & tf3
month_df = month_df[tf].copy()
```

```python
month_df['Current Loan Delinquency Status'].unique()
```

```python
month_df['Current Loan Delinquency Status'] = month_df['Current Loan Delinquency Status'].astype(int)
```

```python
month_df.head()
```

```python
agg_dict = {'Current Loan Delinquency Status':sum, 
            'Monthly Reporting Period':'count'}
delinq_df = month_df.groupby('Loan Sequence Number').agg(agg_dict)
delinq_df.head()
```

```python
col_dct = {'Monthly Reporting Period':'NumMonths'}
delinq_df.reset_index(inplace=True)
delinq_df.rename(columns=col_dct, inplace=True)
delinq_df.head()
```

```python
delinq_df2 = delinq_df[delinq_df['NumMonths']==13].copy()
```

```python
delinq_df2['Loan Deliquency Within Year'] = True

tf = delinq_df2['Current Loan Delinquency Status'] == 0
delinq_df2.loc[tf, 'Loan Deliquency Within Year'] = False
```

```python
delinq_df2['Loan Deliquency Within Year'].value_counts()
```
